```python
# importing matplotlib and seaborn
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
ax = sns.countplot(df, x= "Age Group")
plt.title("Count of Age Group")
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
sns.scatterplot(data=df, x="Age", y="Amount", color="Green", hue="Gender")

```

```python
# plotting all the above 4 plots in a single line with barplot'
plt.figure(figsize = (10,5))
plt.subplot(1,4,1)
df["Gender"].value_counts().plot(kind = "bar")
plt.title(" Gender counts")
plt.subplot(1,4,2)
df["shaadi"].value_counts().plot(kind = "bar")
plt.title(" shaadi counts")
```

```python
sns.countplot(df, x= "Gender")
plt.title("Count of Gender")
```

```python
plt.pie(df["Gender"].value_counts(), labels = ["Female", "Male"], autopct="%.2f%%")
plt.show()
```

```python
sns.countplot(df, x= "shaadi", hue= 'Gender')
plt.title("Count of shaadi")
```

```python
sales_state = df.groupby('State')[['Orders']].sum().reset_index()


import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (10,5))

sns.barplot(data=sales_state, x='State', y='Orders')
plt.title("Total Orders by State")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize = (10,5))

sns.barplot(data=sales_state, x='Occupation')
plt.title("Total Orders by State")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
